## Combine data by month for 5 years 2018 - 2023

```python
# Imports , Pandas
import pandas as pd
```

```python
# List of file names to be combined for 2019
files2019 = ['data/JC-201901-citibike-tripdata.csv', 
             'data/JC-201902-citibike-tripdata.csv', 
             'data/JC-201903-citibike-tripdata.csv', 
             'data/JC-201904-citibike-tripdata.csv', 
             'data/JC-201905-citibike-tripdata.csv', 
             'data/JC-201906-citibike-tripdata.csv', 
             'data/JC-201907-citibike-tripdata.csv', 
             'data/JC-201908-citibike-tripdata.csv', 
             'data/JC-201909-citibike-tripdata.csv', 
             'data/JC-201910-citibike-tripdata.csv', 
             'data/JC-201911-citibike-tripdata.csv', 
             'data/JC-201912-citibike-tripdata.csv']

# Create an empty list to store each dataframe
df2019_list = []

# Loop through each file in the list
for file in files2019:
    # Read the file into a dataframe
    df2019 = pd.read_csv(file)

    # Append the dataframe to the list
    df2019_list.append(df2019)

# Concatenate all dataframes in the list into one dataframe
df2019 = pd.concat(df2019_list)
# Convert the 'starttime and endtime' columns to datetime datatype.
new_start = df2019.starttime
new_stop = df2019.stoptime
# convert objects to datetime
new_start = pd.to_datetime(new_start)
new_stop = pd.to_datetime(new_stop)
# add separate fields to DF for Start and stop dates & times
df2019['StartDate'] = new_start.dt.date 
df2019['StopDate'] = new_stop.dt.date 
df2019['StartTime'] = new_start.dt.time 
df2019['StopTime'] = new_stop.dt.time 
# Show dataframe
df2019
```

```python
# create new df for tableau just the needed columns grouped by date for similar start & end points

df2019_cnt = pd.DataFrame(df2019.groupby(['StartDate', 
                                    'start station id', 
                                    'start station name', 
                                    'start station latitude', 
                                    'start station longitude',	
                                    'end station id',	
                                    'end station name']).size().reset_index(name='count') )
# adding a column for ride type
df2019_cnt ['ride type'] = 'classic_bike'
# Show dataframe
df2019_cnt
```
